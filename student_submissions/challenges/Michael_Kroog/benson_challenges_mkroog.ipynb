```python
import pandas as pd
import datetime
```

```python
def get_mta_data(mta_data):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    df = []
    for mta in mta_data:
        url_file = url.format(mta)
        df.append(pd.read_csv(url_file, parse_dates=[['DATE', 'TIME']], keep_date_col=True))
    return pd.concat(df)
```

```python
mta_data = [180908, 180915, 180922]
turn_df = get_mta_data(mta_data)
```

**Challenge 3**

```python
turn_daily = turn_df.groupby(['C/A', 'UNIT', 'SCP', 'STATION', 'DATE'], as_index = False)['ENTRIES'].first()
```

```python
turn_daily.head()
```

```python
turn_daily[['PREV_DATE', 'PREV_ENTRIES']] = (turn_daily
                                             .groupby(['C/A', 'UNIT', 'SCP', 'STATION'])['DATE', 'ENTRIES']
                                             .apply(lambda x: x.shift(1)))
```

```python
turn_daily.head()
```

```python
turn_daily.dropna(inplace = True)
```

```python
turn_daily[turn_daily['ENTRIES'] < turn_daily['PREV_ENTRIES']].groupby(['C/A', 'UNIT', 'SCP', 'STATION']).size()
```

```python
def get_daily_counts(row, max_counter):
    counter = row['ENTRIES'] - row['PREV_ENTRIES']
    if counter < 0:
        counter = -counter
    if counter > max_counter:
        print(row['ENTRIES'], row['PREV_ENTRIES'])
        counter = min(row['ENTRIES'], row['PREV_ENTRIES'])
    if counter > max_counter:
        return 0
    return(counter)
turn_daily['DAILY_ENTRIES'] = turn_daily.apply(get_daily_counts, axis = 1, max_counter=86400)
```
