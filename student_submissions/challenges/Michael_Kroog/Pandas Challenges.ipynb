**Challenge 1**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
from scipy.stats import sem
import numpy as np
```

```python
df = pd.read_csv('2013_movies.csv', parse_dates = {'Date': ['ReleaseDate']})
```

```python
df = df.sort_values(['Date'])
plt.figure(figsize=(15,5))
plt.plot(df['Date'], df['DomesticTotalGross'])
plt.xlabel('Release Date')
plt.ylabel('Gross Domestic Total')
plt.title('Gross Domestic Total (2013)')
```

**Challenge 2**

```python
df = df.sort_values(['Runtime'])
plt.figure(figsize=(15,5))
plt.plot(df['Runtime'], df['DomesticTotalGross'])
plt.xlabel('Runtime')
plt.ylabel('Gross Domestic Total')
plt.title('Gross Domestic Total (2013)')
```

**Challenge 6**

```python
df_sem = df.groupby('Month')['DomesticTotalGross'].agg(sem).reset_index()
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df_month = df.groupby('Month')['DomesticTotalGross'].mean().reset_index()
```

```python
plt.bar(df_month['Month'], df_month['DomesticTotalGross'], yerr = df_sem['DomesticTotalGross'])
plt.title('Mean Domestic Total Gross by Month in 2013')
plt.ylabel('Domestic Total Gross')
plt.xlabel('Month')
```
