```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('2013_movies.csv')
```

```python
df
```

## CHALLENGE 1

```python
df['ReleaseDate'] = pd.to_datetime(df['ReleaseDate'])
```

```python
df.plot(x="ReleaseDate", y="DomesticTotalGross")
```

## CHALLENGE 2

```python
plt.figure(figsize=(20, 5))
df.sort_values('Runtime', inplace = True)
df.plot(x="Runtime", y="DomesticTotalGross")
```

## CHALLENGE 3

```python
rating_grouped_df = df.groupby(['Rating'])["Runtime", "DomesticTotalGross"].mean()
```

```python
rating_grouped_df
```

## CHALLENGE 4

```python
g_plot = df.loc[df["Rating"] == 'G']
pg_plot = df.loc[df["Rating"] == 'PG']
pg13_plot = df.loc[df["Rating"] == 'PG-13']
r_plot = df.loc[df["Rating"] == 'R']
```

```python

plt.figure(figsize=[30,30])
plt.suptitle('Movie Ratings(Release Date vs Domestic gross)',fontsize = 40)
plt.subplot(2,2,1)
plt.plot(g_plot["ReleaseDate"].sort_values(), g_plot["DomesticTotalGross"])
plt.title("G Rating", fontsize=10)
plt.xticks(fontsize=20)
plt.subplot(2, 2, 2)
plt.plot(pg_plot["ReleaseDate"].sort_values(), pg_plot["DomesticTotalGross"])
plt.title("PG Rating", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.subplot(2, 2, 3)
plt.plot(pg13_plot["ReleaseDate"].sort_values(), pg13_plot["DomesticTotalGross"])
plt.title("PG-13 Rating", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.subplot(2, 2, 4)
plt.plot(r_plot["ReleaseDate"].sort_values(), r_plot["DomesticTotalGross"])
plt.title("R Rating", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
```

## CHALLENGE 6

```python
df["Months"] = df["ReleaseDate"].dt.month_name()
```

```python
mean_month_df = df.groupby(["Months"])["DomesticTotalGross"].mean().sort_values().reset_index()
```

```python
mean_month_df
```

```python
plt.figure(figsize=[15,15])
plt.title("Mean Domestic Total Gross by Month in 2013")
plt.bar(mean_month_df["Months"], mean_month_df["DomesticTotalGross"])
```
