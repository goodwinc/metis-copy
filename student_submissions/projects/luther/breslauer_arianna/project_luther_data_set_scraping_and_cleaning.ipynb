```python
import requests
import pandas as pd
import numpy as np
import re
from datetime import datetime
from bs4 import BeautifulSoup
import sys
import pickle
```

```python
# Function to read each url
def read_url_imdb(url):
    response = requests.get(url)
    page = response.text
    #soup = BeautifulSoup(page, "lxml")
    soup = BeautifulSoup(page, 'lxml')
    return soup
```

```python
def read_url_boxofficemojo(url):
    response = requests.get(url)
    page = response.text
    #soup = BeautifulSoup(page, "lxml")
    soup = BeautifulSoup(page, 'xml')
    return soup
```

```python
# Using 100 pages of 50 movies per page, 5000 movies, sorted by IMDB popularity
url_list_imdb=[]
for i in range(1,101):
    url_list_imdb.append('https://www.imdb.com/search/title?genres=comedy&explore=title_type,genres&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=75c37eae-37a7-4027-a7ca-3fd76067dd90&pf_rd_r=B51V9XKGTKPP1B8G54N0&pf_rd_s=center-1&pf_rd_t=15051&pf_rd_i=genre&title_type=movie&page={}&ref_=adv_nxt'.format(i))
```

```python
# Confirming length of list is 100 as 100 pages
len(url_list_imdb)
```

```python
# Creating a list of the titles of the top IMDB movies
movie_titles_imdb = []
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    for entry in data_movies_imdb.find_all("div", {"class": "lister-item-content"}):
        movie_title = entry.find('a').contents[0]
        movie_titles_imdb.append(movie_title)
```

```python
# Collecting IMDB ratings for movies from IMDB
imdb_ratings=[]
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    for entry in data_movies_imdb.find_all("div",{"class": "lister-item-content"}):
        try:
            if entry.find('div', {"class":"ratings-bar"}):
                rating = float(entry.find('div',{'class':'ratings-bar'}).find('strong').get_text())
                imdb_ratings.append(rating)
            else:
                imdb_ratings.append(np.nan)
        except:
            imdb_ratings.append(np.nan)
```

```python
# Collecting MPAA ratings for movies from IMDB
MPAA_ratings=[]
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    for entry in data_movies_imdb.find_all("div", {"class": "lister-item-content"}):
        try: 
            if entry.find('p', {'class':'text-muted '}).find('span',{'class': 'certificate'}):
                mpaa_rating = entry.find('p', {'class':'text-muted '}).find('span',{'class': 'certificate'}).get_text()
                MPAA_ratings.append(mpaa_rating)
            else:
                MPAA_ratings.append('Not Available')
        except:
            MPAA_ratings.append('Not Available')
```

```python
# Collecting genres for movies from IMDB
genres_imdb=[]
for url in url_list_imdb:
    genre=read_url_imdb(url).find_all('span',{"class":"genre"})
    for item in [g.get_text() for g in genre]:
        genres_imdb.append(str(item).replace('\n','').strip())
```

```python
# Find no. of unique genres and combine 'Music' and 'Musical' categories
from collections import defaultdict
unique_genres_imdb = []
for entry in genres_imdb:
        entry_list = str(entry).replace(' ','').replace('Musical','Music').replace('-','').split(',')
        for genre in entry_list:
            if genre not in unique_genres_imdb:
                unique_genres_imdb.append(genre)
unique_genres_imdb = sorted(unique_genres_imdb)
print(unique_genres_imdb)
```

```python
# Converting genres to dummy variable
genre_set=['Action','Adventure','Animation','Biography','Comedy','Crime','Drama','Family','Fantasy','History','Horror',
      'Music','Mystery','Romance','SciFi','Sport','Thriller','War','Western']
Action,Adventure,Animation,Biography,Comedy,Crime,Drama,Family,Fantasy,History,Horror,Music,Mystery,Romance,SciFi,Sport,Thriller,War,Western=([] for i in range(len(genre_set)))
# Creates empty list for each genre represented in the data

genre_list=[Action,Adventure,Animation,Biography,Comedy,Crime,Drama,Family,Fantasy,History,Horror,Music,Mystery,Romance,SciFi,Sport,Thriller,War,Western]
for g in genres_imdb:
    for c in genre_set:
        if c in g:
            genre_list[genre_set.index(c)].append(1)
        else:
            genre_list[genre_set.index(c)].append(0)
```

```python
# Collecting runtimes for movies from IMDB
runtimes_imdb = []
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    for entry in data_movies_imdb.find_all("p", {"class": "text-muted"}):
        if entry.find('span')==None:
            pass
        elif entry.find_all('span')[0]['class']==['certificate'] and entry.find_all('span')[2]['class']!=['genre']:
            runtime_digits = re.findall(r"\d+", entry.find_all('span')[2].get_text()).pop()
            runtimes_imdb.append(int(runtime_digits))
        elif entry.find_all('span')[0]['class']==['runtime']:
            runtime_digits = re.findall(r"\d+", entry.find_all('span')[0].get_text()).pop()
            runtimes_imdb.append(int(runtime_digits))
        else:
            runtimes_imdb.append(np.nan)
```

```python
# Find budgets from IMDB 
# First get each movie id for their url
movie_urls_imdb = []
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    for entry in data_movies_imdb.find_all("div",{"class": "lister-item-content"}):
        try: 
            if entry.find('h3', {'class':'lister-item-header'}).find('span',{'class': 'lister-item-index unbold text-primary'}):
                movie_url_imdb = entry.find('h3', {'class':'lister-item-header'}).find('a')['href']
                movie_urls_imdb.append(movie_url_imdb)
            else:
                movie_urls_imdb.append(np.nan)
        except:
            movie_urls_imdb.append(np.nan)
```

```python
individual_movie_urls_imdb = ['https://www.imdb.com' + i for i in movie_urls_imdb]
```

```python
movie_budgets= []
i = 1
for url in individual_movie_urls_imdb:
    data_movies_imdb = read_url_imdb(url)
    try:
        if data_movies_imdb.find(id="titleDetails").find_all('div', {'class':'txt-block'})[6]:
            budget_table = data_movies_imdb.find(id="titleDetails").find_all('div', {'class':'txt-block'})[6]
            budget_table = budget_table.get_text().replace('$','').replace(',','')
            budget = re.sub('\D','',budget_table)
            i += 1
            movie_budgets.append(budget)
            if i % 100 == 0:
                print(budget)
                with open('movie_budgets.pkl', 'wb') as f:
                    pickle.dump(movie_budgets, f)
        else:
            movie_budgets.append(np.nan)
    except:
        movie_budgets.append(np.nan)
# the ones without budgets are 2018/2019 films - will drop later as just doing 2018
```

```python
# Scraping actors for each movie
movie_actors_imdb = []
for url in url_list_imdb:
    data_movies_imdb = read_url_imdb(url)
    tables = data_movies_imdb.find_all('div', {"class":"lister-item mode-advanced"})
    for table in tables:
        actors = []
        if table.find(text=re.compile("Stars:")):
            stars = table.find(text=re.compile("Stars:"))
            actors.append(stars.findNext().get_text().replace(',','').replace('.','').replace("'",'').replace('-',''))
            actors.append(stars.findNext().findNext().get_text().replace(',','').replace('.','').replace("'",'').replace('-',''))            
            actors.append(stars.findNext().findNext().findNext().get_text().replace(',','').replace('.','').replace("'",'').replace('-',''))         
            actors.append(stars.findNext().findNext().findNext().findNext().get_text().replace(',','').replace('.','').replace("'",'').replace('-',''))
        elif table.find(text=re.compile("Star:")):
            star = table.find(text=re.compile("Star:"))
            actors = star.findNext().get_text().replace(',','').replace('.','').replace("'",'').replace('-','')
        else:
            actors = 'TBD'
        print(actors)
        movie_actors_imdb.append(actors)
```

```python
# Acquiring movies A-Z from BoxOfficeMojo.com
boxofficemojo_list=[]
boxofficemojo_dict={'A':10,'B':8,'C':7,'D':6,'E':7,'F':7,'G':7,'H':6,'I':6,'J':4,'K':4,'L':5,'M':7,'N':5,'O':5,'P':8,'Q':1,'R':6,'S':13,'T':8,'U':2,'V':3,'W':6,'X':1,'Y':2,'Z':2}
for letter in boxofficemojo_dict.keys():
    for num in range(1,boxofficemojo_dict[letter]+1):
        boxofficemojo_list.append('http://www.boxofficemojo.com/movies/alphabetical.htm?letter={}&page={}&p=.htm'.format(letter,num))
```

```python
titles,dates,opening_gross,opening_theaters,total_US_gross,movie_ids= ([] for i in range(6))
for url in boxofficemojo_list: 
    data_movies_mojo = read_url_boxofficemojo(url)
    if len(data_movies_mojo.find_all("table")) == 5:
        table1 = data_movies_mojo.find_all("table")[3]
        rows = table1.find_all('tr')
        for row in rows[1:]:
            try:
                if row.find_all('td')[0].find('a').get_text():
                    titles.append(str(row.find_all('td')[0].find('a').get_text()))
                else:
                    titles.append(np.nan)
            except:
                titles.append(np.nan)
            try:
                if row.find_all('td')[4]:
                    opening_gross.append(int(str(row.find_all('td')[4].get_text()).replace('$','').replace(',','').replace('n/a','0')))
                else:
                    opening_gross.append(np.nan)
            except:
                opening_gross.append(np.nan)
            try:
                if row.find_all('td')[2].get_text():
                    total_US_gross.append(int(str(row.find_all('td')[2].get_text()).replace('$','').replace(',','').replace('*','').replace('n/a','0')))
                else:
                    total_US_gross.append(np.nan)
            except:
                total_US_gross.append(np.nan)
            try:
                if row.find_all('td')[6]:
                    dates.append(str(row.find_all('td')[6].get_text()))
                else:
                    dates.append(np.nan)
            except:
                dates.append(np.nan)
            try:
                if row.find_all('td')[5]:
                    opening_theaters.append(int(row.find_all('td')[5].get_text().replace(',','')))
                else:
                    opening_theaters.append(np.nan)
            except:
                opening_theaters.append(np.nan)
            try:
                if row.find_all('td')[0].find('a'):
                    movie_ids.append(row.find_all('td')[0].find('a')['href'])
                else:
                    movie_ids.append(np.nan)
            except:
                movie_ids.append(np.nan)
```

```python
movie_id_urls = ['https://www.boxofficemojo.com' + item.replace('?id=','?page=intl&id=') for item in movie_ids]    
```

```python
uk_total_gross = []
import re
for url in movie_id_urls:
    data_movies_mojo = read_url_boxofficemojo(url)
    dist_regex = data_movies_mojo.find(text='Dist.')
    uk_gross_regex = re.compile('United Kingdom')
    uk_title = data_movies_mojo.find(text=uk_gross_regex)
    try:
        if data_movies_mojo.find(text=uk_gross_regex):
            if dist_regex:
                uk_gross = uk_title.findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().get_text().replace('$','').replace(',','').replace('*','').replace('n/a','0')
            else:
                uk_gross = uk_title.findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().findNext().get_text().replace('$','').replace(',','').replace('*','').replace('n/a','0')
        else:
            uk_gross = np.nan
    except:
        uk_gross = np.nan
    uk_total_gross.append(uk_gross)
print(uk_total_gross)
```

```python
ex_US_grosses = []
for url in movie_id_urls2:
    data_movies_mojo = read_url_boxofficemojo(url)
    ex_US_regex = data_movies_mojo.find(text='Foreign:')
    try:
        ex_US_gross = ex_US_regex.findNext().get_text().replace('$','').replace(',','').replace('*','').replace('n/a','0')
    except:
        ex_US_gross = np.nan
    if ex_US_gross ==0:
        ex_US_gross = np.nan
    ex_US_grosses.append(ex_US_gross)
    if len(ex_US_grosses) % 100:
        print(ex_US_gross)
```
