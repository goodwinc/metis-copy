```python
urlsOfInterest = ['http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2002',
                 'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2004',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2005',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2006',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2007',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2008',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2009',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2010',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2011',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2012',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2013',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2014',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2015',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2016',
                  'http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2017'
                 ]

driver = webdriver.Chrome(chromedriver)
#driver.get("http://www.espn.com/mens-college-basketball/statistics/player/_/stat/scoring-per-game/sort/avgPoints/year/2002/seasontype/3")

```

```python
content_url_list = []
```

```python
for urls in urlsOfInterest:
    driver.get(urls)
    for i in range (0, 3):
        content_url = driver.current_url
        content_url_list.append(content_url)
        next_button = driver.find_element_by_class_name("jcarousel-next")
        next_button.click()
```

```python
print(content_url_list)
```

```python
import pandas as pd

# list of dictionaries is a very nice parsing format: pandas
# can automatically convert it to a dataframe
headers = ['Rank','Player', 'Team', 'Games Played', 'Minutes Per Game', 'Points Per Game', 'FGM-FGA', 'FG%', '3PM-3PA',
           '3P%', 'FTM-FTA', 'FT%']

big_list = []      
for urls in content_url_list:
    driver.get(urls)
    x = driver.find_elements_by_class_name("oddrow")
    #y = driver.find_elements_by_class_name("evenrow")
    for item in x:
        intermediateList = []
        for i in range (0, 12):
            intermediateList.append(item.find_elements_by_tag_name('td')[i].text)
        big_list.append(intermediateList)
        

for urls in content_url_list:
    driver.get(urls)
    y = driver.find_elements_by_class_name("evenrow")
    #y = driver.find_elements_by_class_name("evenrow")
    for item in y:
        intermediateList = []
        for i in range (0, 12):
            intermediateList.append(item.find_elements_by_tag_name('td')[i].text)
        big_list.append(intermediateList)
```

```python
college_df_regular_season = pd.DataFrame(big_list, columns = headers)
```
