## Price regression analyis of appartments in Budapest, Hungary

---

**1. Objective:** Predict offer prices of Budapest apartments based on standard features (e.g.  floor, bedrooms, location) and derived features (e.g. distance from city centre) based on data scraped from ingatlan.com.

**2. Baseline model:**

![](SIZE-PRICE-BASELINE.png)

The figure depicts the baseline regression model between SIZE as the independent, and PRICE as the dependent variable. The model has an R-SQUARED of ~64%, and an RMSE of ~19 million HUF (~67k USD). This result suggests that SIZE is not the only important factor in predicting the appartment PRICE, and there are numerous optimization factors to be implemented. Below I describe the brief methodology of my next steps.

**3. Next steps:**
1. Clean the available feature and target set from from outliers, and NaNs
2. Apply one-hot encoding on the features where applicable, e.g. district, presence of an elevator, type of view, orientation of living room, etc.
3. Add further cleaned numerical and one-hot-encoded features to the regression model and fit linear models
4. Add 2nd degree and interaction numerical variables to the model to account for polynomial relationship with the target variable
5. Apply LASSO regularization on the resulting model to limit the number of variables generated by the 2nd degree and interaction numerical variables
