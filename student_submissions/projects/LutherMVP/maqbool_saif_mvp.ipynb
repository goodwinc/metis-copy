**Cult Classics: Minimum Viable Product**

```python
% matplotlib inline
import pickle
import pandas as np
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt

cine = pickle.load( open( "truffaut.pkl", "rb" ) )
```

In setting out to effectively predict films that go on to enjoy 'cult' status, initial engagement with the dataset has begun to reveal promising avenues for analysis and the suitability of this data for a linear regression model. Cult hits by our definition are films that accrue the majority of their revenue after the opening weekend of their domestic release.

```python
np.mean(cine['opening_weekend_percentage'])
```

As we can see, films on average tend to amass 26.7% of their total domestic gross on the opening weekend. By predicting this percentage, we will be able to project how much films will go on to earn after their initial release, thereby allowing for us to maximize this afterburner effect through a more expansive and meticulous advertising campaign. 

```python
sns.distplot(cine['opening_weekend_percentage'])
```

As we may observe, opening weekend as a percentage of domestic gross has a normal distribution, which bodes well for the success of our model, which assumes such a distribution as one of its core premises. A broader analysis of the dataset using seaborn's helpful pairplot function reveals more: 

```python
sns.pairplot(cine)
```
