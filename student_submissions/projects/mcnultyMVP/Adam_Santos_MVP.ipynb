# McNutty Minimum Viable Product 10/25

My project is to create a prediction system for the website Donor's Choose. This website is a platform to connect teachers who want funding for classroom projects with people who would like to donate to teachers. I would like to create a system that predicts whether or not a project will receive full funding when it is posted. If they project is unlikely to receive full funding, then the system will give the teacher recommendations on what to change about the project.

My first step was to look at the distribution of completed projects and if they received funding or not. This is shown in the bar chart below.

Around 77 percent of projects receive full funding.

My next step was to run a basic logistic regression model with a few features. I got an accuracy of 0.78 and an F1 score of 0.09. This about on par with always predicting that a project will be fully funded so there is some work to be done to find better ways to predict whether a project will be funded or not. My next steps are to add more features from available information and possibly try more sophisticated models.
