```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
```

```python
dating1=pd.read_csv('/Users/andreewerner/Downloads/SpeedDatingDataCleaned.csv')
dating1;
```

```python
sns.lmplot("attr1_1", "amb1_1",
           data=dating1, hue='gender', fit_reg=False,
           x_jitter=5, y_jitter=5,
           palette=['#F08080', '#6495ED'],
           scatter_kws={'alpha':.03})
plt.title('how men and women rate importance of attractiveness vs. ambition')
```

```python
sns.kdeplot(dating1[dating1['gender'] == 1].age,
            bw=1.5, shade=True, color='#6495ED', label='men')
sns.kdeplot(dating1[dating1['gender'] == 0].age,
            bw=1.5, shade=True, color='#F08080', label='women')
plt.title('Age of daters')
plt.xlabel('Age')
```

```python
sns.kdeplot(dating1[(dating1['gender'] == 1) & (dating1['income'] > 0)]['income'],
            bw=15000, shade=True, color='#6495ED', label='men')
sns.kdeplot(dating1[(dating1['gender'] == 0) & (dating1['income'] > 0)]['income'],
            bw=15000, shade=True, color='#F08080', label='women')
plt.title('Estimated income of daters')

plt.xlim(0,150000)
plt.xlabel('Zip code median income')
```
