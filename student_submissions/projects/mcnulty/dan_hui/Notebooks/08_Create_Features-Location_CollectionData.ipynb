This notebook creates features from the collection dataset

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

#### Load Inventory and only keep relevant Columns

```python
inventory_df = pd.read_csv('../01_Data/03_Cleaned/Library_Collection_Inventory_jan_2018_clean.csv',index_col=0)
```

```python
inventory_df.head()
```

```python
location_df = inventory_df[["BibNum","ItemLocation"]]        #Just keep the library locations
location_df = location_df.drop_duplicates()                  #remove duplicates (a single book in multiple 
location_df.head()                                           #locations within the same branch
```

```python
len(location_df)    #992,518 Records
```

#### Make Dummy List of all Branch Locations

```python
location_dummies_df = pd.get_dummies(location_df["ItemLocation"])   #branch dummies locations
len(location_dummies_df)                                            #992,518 Records
```

```python
location_dummies_df.head()
```

#### Merge DataFrames together to create feature table

```python
location_features_df = pd.merge(location_df,location_dummies_df,left_index=True,right_index=True)
```

```python
location_features_df = location_features_df.groupby(["BibNum"]).sum()
location_features_df = location_features_df.reset_index()
location_features_df.tail(5)
```

```python
len(location_features_df)          #397,147 Unique Titles
```

### Export Data

```python
location_features_df.to_csv("../01_Data/06_Features/Location_Features.csv")
```
