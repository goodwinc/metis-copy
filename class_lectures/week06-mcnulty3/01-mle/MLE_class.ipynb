**The Big IDEA**

- So far, when we have solved for the best parameters for our models (Betas) we have done
so by attempting to find the beta that provided the best fit of our data. 


- **Idea with MLE**: maximize the probability of the parameters given the data. 
(seek the parameters that most likely produced the data) 

##### Suppose we have some data: 
    y = [y1,y2,y3, y4...  ]
    X = [[x11, x21, x31,.. ],[x12,22,x32..]]
    


##### GOAL:  We want to identify the population that most likely generated this sample. 

In stats, a population can be identified by a probability distribution.
Associated with each probability distribution is a unique value of 
parameters.   *As these parameters change, different probability 
distributions are generated*

Let's say we have f(y|w) that is our pdf:  that specifies the probability of observing data vector y, given parameter w.

Lets start with a simple example : 

####  10 Bernouli trials: 

$f(y|n =10, w= 0.2) = {{10!\above 1pty!(10-y)!}}(0.2)^y(.8)^{10-y}$

```python
def binomial(n,p,y):
    func=p**y*(1-p)**(n-y)*math.factorial(n)/(math.factorial(y)*math.factorial(n-y))
    return func
```

```python
# Example:  10 Bernoulli trials, p =0.2
N=10
p=0.2

# f(y| n=10,w=0.2)

outcomes=[]

for i in range(11):
    outcomes.append(binomial(10,0.2,i))
```

```python
test=[i/10 for i in outcomes]
plt.bar(range(11),test,align='center')

plt.xlabel('y')
plt.ylabel('f(y|n=10,p=.2)')
```

```python
# Example:  10 Bernoulli trials, p =0.7
outcomes=[]

for i in range(11):
    outcomes.append(binomial(10,0.7,i))
```

```python
plt.bar(range(11),outcomes)
plt.xlabel('y')
plt.ylabel('f(y|n=10,p=.7)')
```

**We see that:**
Given a set of parameter values (s.a. some value of p:), the corresponding PDF will show that some data are more probable than other data.  

**What about the inverse problem ? **
*In reality, we already have the y data --> we want to calculate the p*     
Given the observed data and a model of interest -> find the PDF (in this case, the probability parameter)
that is most likely to have produced this data.  To solve this inverse problem:  Let's define a liklihood function by reversing the roles of the data vector y & the parameter vector w in f(y|w). 

Let's find Liklihood: 
$L(w|n=10, y=7)$ 

**We will see that:**
$L(w|n=10, y=7) = f(y| n=10,w= .7)$ 

```python
# = $(10!/7!3!)w^7(1-w)^3$


outcomes=[]

for i in np.arange(0,1,.01):
    outcomes.append(binomial(10,i,7))
    
plt.plot(np.arange(0,1,.01),outcomes)
plt.xlabel('parameter p')
plt.ylabel('Liklihood')
```

I.e :  Given some value of y (y=7), 
Liklihood(for y=7) = probability distribution(for p=.7)


**Likelihood:**
represents the likelihood of parameter p, given some observed data y
