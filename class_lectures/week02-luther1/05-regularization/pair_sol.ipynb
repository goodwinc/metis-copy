**Plotting The Surface**

```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def f(x,y): 
    return (x**2-8*x+7)/3 + (y**2-5*x+8)/6

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')
```

**Gradient Descent**

First calculate the partial derivatives to get the answers below

```python
def df_x(x,y):
    return (2*x - 8)/3 - 5/6

def df_y(x,y):
    return (y/3)
```

Now write gradient descent function to iteratively update x and y along direction of gradient with step size of rate

```python
def descent(rate, iterations):
    x, y = 12, 5
    for i in range(iterations):
        x = x - rate * df_x(x,y)
        y = y - rate * df_y(x,y)
    return x, y
```

```python
descent(.01, 10000000)
```
